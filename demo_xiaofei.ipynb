```python
# 异常值处理 分位数计算法
def outlier_processing(dfx):
    df = dfx.copy()
    q1 = df.quantile(q=0.25)
    q3 = df.quantile(q=0.75)
    iqr = q3 - q1
    Umin = q1 - 1.5*iqr
    Umax = q3 + 1.5*iqr 
    df[df>Umax] = df[df<=Umax].max()
    df[df<Umin] = df[df>=Umin].min()
    return df
```

```python
data['age_month']=outlier_processing(data['age_month'])#处理异常值
plt.figure()#重画age_month的分布图 
sns.boxenplot(x='result', y=u'age_month', data=data)
plt.show()
```

# 特征提取

```python
str_features = []
num_features=[]
for col in data.columns:
    if data[col].dtype=='object':
        str_features.append(col)
        print(col,':  ',data[col].unique())
    if data[col].dtype=='int64' and col not in ['user_id']:
        num_features.append(col)
print(str_features)
print(num_features)# 分离离散值和连续值
```

```python
#----------数据集处理--------------
from sklearn.model_selection import train_test_split        # 划分训练集和验证集
from sklearn.linear_model import LogisticRegression         # 逻辑回归

#----------数据处理--------------
from sklearn.preprocessing import StandardScaler # 标准化
```

```python
data_new=data.join(pd.get_dummies(data[str_features].city_num)).drop(str_features,axis=1)#独热编码再进行训练集划分
```

```python
data_new
```

# 划分数据集

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

y = data_new['result']
x = data_new.drop(['result', 'user_id'], axis=1)
# 数据分割
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)

# 标准化
std = StandardScaler()
x_train = std.fit_transform(x_train)
print(x_train)
x_test = std.fit_transform(x_test)
```

# 模型构建与预测评估

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
y_predict = lr.predict(x_test)
print("得出来的权重：", lr.coef_)
print("预测的类别：", lr.predict(x_test))

 # 得出准确率
print("预测的准确率:", lr.score(x_test, y_test))
```

```python
# 考察精确率、召回率和F1-score
from sklearn.metrics import classification_report
report = classification_report(y_test, y_predict, labels=[1,0], target_names=["购买", "未购买"] )
print(report)
```

```python
# 样本不均衡，考察AUC指标
from sklearn.metrics import roc_auc_score
print("AUC指标：", roc_auc_score(y_test, lr.predict(x_test)))
```

```python
# 绘制ROC曲线
def get_rocauc(X,y,clf):
    from sklearn.metrics import roc_curve
    FPR,recall,thresholds=roc_curve(y,clf.predict_proba(X)[:,1],pos_label=1)
    area=roc_auc_score(y,clf.predict_proba(X)[:,1])
    maxindex=(recall-FPR).tolist().index(max(recall-FPR))
    threshold=thresholds[maxindex]
    plt.figure()
    plt.plot(FPR,recall,color='red',label='ROC curve (area = %0.2f)'%area)
    plt.plot([0,1],[0,1],color='black',linestyle='--')
    plt.scatter(FPR[maxindex],recall[maxindex],c='black',s=30)
    plt.xlim([-0.05,1.05])
    plt.ylim([-0.05,1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('Recall')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc='lower right')
    plt.show()

get_rocauc(x_test,y_test,lr)
```
